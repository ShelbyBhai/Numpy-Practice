```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(5)
print(x)
y = x
print(y)
```

```python
plt.plot(x,y,'o--')
plt.plot(-x,y,'o-')
plt.title('y=abs(x)')
plt.show()
```

```python
x = np.linspace(0,12,7)
print(x)
#y = x
y = np.logspace(np.euler_gamma,np.e,7)
print(y)
plt.plot(y,x,'o')
#plt.axis('off')
plt.show()
z = np.geomspace(np.pi,1000,7)
print(z)
plt.plot(x,z,'o-')
plt.show()
plt.plot(y,z,'o--')
plt.show()
```
